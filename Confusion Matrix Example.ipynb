```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
y_true = np.array(['dog', 'cat', 'cat', 'dog', 'dog'])
y_pred = np.array(['dog', 'cat', 'cat', 'cat', 'dog'])
```

```python
confusion_matrix = confusion_matrix(y_true, y_pred, labels=['cat', 'dog'])
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["cat", "dog"])
```

```python
cm_display.plot()
```
